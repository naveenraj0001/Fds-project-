```python
import numpy as np


arr = np.random.randint(1, 100, 16)
arr
```

```python
arr.mean()
```

```python
np.percentile(arr, 25)
```

```python
np.percentile(arr, 50)
```

```python
np.percentile(arr, 75)
```

```python
np.percentile(arr, 100)
```

```python
def out_detect(arr):
    arr.sort()
    Q1, Q3 = np.percentile(arr, [25, 75])
    IQR = Q3 - Q1
    lr = Q1 - (1.5 * IQR)
    ur = Q3 + (1.5 * IQR)
    return lr, ur

lr, ur = out_detect(arr)
lr, ur
```

```python
import seaborn as sns

sns.displot(arr)
```

```python
sns.distplot(arr)
```

```python
new_arr = arr[(arr > lr) & (arr < ur)]
new_arr
```

```python
sns.displot(new_arr)
```

```python
lrl, url = out_detect(new_arr)
lrl, url
```

```python
final_arr = new_arr[(new_arr > lr) & (new_arr < ur)]
final_arr
```

```python
sns.distplot(final_arr)
```
